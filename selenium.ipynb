```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
driver =  webdriver.Chrome()
```

```python
driver.get("https://www.youtube.com")
```

```python
time.sleep(2)
search_text = input('What would you want to search here --> ')
search_box = driver.find_element(By.XPATH, '//*[@id="center"]/yt-searchbox/div[1]/form/input')
search_box.click()
search_box.send_keys(search_text)
search_box.send_keys(Keys.RETURN)
```

```python
titles = driver.find_elements(By.XPATH, '//*[@id="video-title"]/yt-formatted-string')
```

```python
with open("Youtube_names.txt", "w", encoding="utf-8") as f:
    for title in titles:
        f.write(title.text + "\n") 
```
